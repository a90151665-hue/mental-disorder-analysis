```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
df["Expert Diagnose"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Distribution of Expert Diagnoses")
plt.xlabel("Diagnosis")
plt.ylabel("Number of Patients")
plt.show()
```

```python
import seaborn as sns
```

```python
df[["Sexual Activity", "Concentration"]].plot(
    kind="hist", bins=5, alpha=0.7, figsize=(7,4))
plt.title("Distribution of Psychological Scores")
plt.xlabel("Score (Out of 10)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Mood Swing", hue="Expert Diagnose", data=df)
plt.title("Mood Swing Frequency by Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```
